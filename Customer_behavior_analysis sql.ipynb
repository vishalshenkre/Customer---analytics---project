```python
import pandas as pd

df = pd.read_csv("marketing_campaign_clean.csv")
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Income'].fillna(df['Income'].median(), inplace=True)
```

```python
df['Age'] = 2025 - df['Year_Birth']
```

```python
def age_group(age):
    if age < 18:
        return 'Minor'
    elif age <= 24:
        return 'Young Adult'
    elif age <= 59:
        return 'Adult'
    else:
        return 'Senior'

df['Age_Group'] = df['Age'].apply(age_group)
```

```python
final_cols = [
    'ID',
    'Age',
    'Age_Group',
    'Education',
    'Marital_Status',
    'Income',
    'Total_Children',
    'Total_Spending',
    'NumWebPurchases',
    'NumStorePurchases',
    'NumCatalogPurchases',
    'Response'
]

df_final = df[final_cols]
df_final.shape
```

```python
df['Total_Spending'] = (
    df['MntWines'] +
    df['MntFruits'] +
    df['MntMeatProducts'] +
    df['MntFishProducts'] +
    df['MntSweetProducts'] +
    df['MntGoldProds']
)
```

```python
df['Total_Children'] = df['Kidhome'] + df['Teenhome']
```

```python
df_final.groupby('Age_Group')['Total_Spending'].mean().sort_values(ascending=False)
```

```python
df_final.groupby('Age_Group')['Response'].mean()
```
